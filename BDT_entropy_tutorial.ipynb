```python
import numpy as np

```

```python
def entropy(y):

    # Empty node
    if len(y) == 0:
        return 0

    # Fraction of signal
    p1 = np.sum(y == 1) / len(y)

    # Fraction of background
    p0 = np.sum(y == 0) / len(y)

    H = 0

    if p1 > 0:
        H -= p1 * np.log2(p1)

    if p0 > 0:
        H -= p0 * np.log2(p0)

    return H




# Information Gain

def information_gain(parent, left, right):

    n = len(parent)

    n_left = len(left)
    n_right = len(right)

    if n_left == 0 or n_right == 0:
        return 0

    H_parent = entropy(parent)
    H_left = entropy(left)
    H_right = entropy(right)

    H_children = (n_left/n)*H_left + (n_right/n)*H_right

    return H_parent - H_children



#-------------------------------------------------------
# Search every feature and every threshold
#-------------------------------------------------------

def best_split(X, y, feature_names):

    best_gain = -1
    best_feature = None
    best_threshold = None

    n_features = X.shape[1]

    for feature in range(n_features):

        # Sort unique values
        values = np.sort(np.unique(X[:, feature]))

        # Midpoints between neighbouring values
        thresholds = (values[:-1] + values[1:]) / 2

        for threshold in thresholds:

            left_mask = X[:, feature] < threshold
            right_mask = ~left_mask

            left_y = y[left_mask]
            right_y = y[right_mask]

            gain = information_gain(y, left_y, right_y)

            print(
                feature_names[feature],
                " threshold =",
                threshold,
                " gain =",
                round(gain,4)
            )

            if gain > best_gain:

                best_gain = gain
                best_feature = feature
                best_threshold = threshold

    print("\nBest split")
    print("Feature :", feature_names[best_feature])
    print("Threshold :", best_threshold)
    print("Information Gain :", best_gain)

    return best_feature, best_threshold






```

```python
###############################################################
# Read the data
###############################################################

events=[]

with open("/content/sample_data/BDT_data.txt") as f:

    next(f)

    for line in f:

        c=line.split()

        pt=float(c[1])
        r9=float(c[2])

        label=1 if c[3]=="S" else 0

        events.append((pt,r9,label))

X=np.array([(e[0],e[1]) for e in events])
print(X)

y=np.array([e[2] for e in events])

feature_names=["pT","R9"]
```

```python

feature,threshold = best_split(X,y,feature_names)
```
